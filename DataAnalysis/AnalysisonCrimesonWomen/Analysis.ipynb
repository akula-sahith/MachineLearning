```python
#Importing the libraries required
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WomenCrimeData.csv')
```

```python
df
```

```python
df.head(5)
```

```python
year_wise_crimes = df.groupby('YEAR')['TOTAL CRIMES AGAINST WOMEN'].sum()
year_wise_crimes.plot(kind='bar',color='red',xlabel='year',ylabel='Number of crimes',title='Year wise crimes')
```

```python
year_wise_rapes = df.groupby('YEAR')['RAPE'].sum()
year_wise_rapes.plot(kind='bar',xlabel='Year',ylabel='Number of rapes')
```

```python
df.head(0)
```

```python
#State wise crime analysis 
state_wise_crimes = df.groupby('STATES')['TOTAL CRIMES AGAINST WOMEN'].sum()
plt.figure(figsize=(15,15))
state_wise_crimes.plot(kind='bar',xlabel="State",ylabel='Number of crimes')
```

```python
df.head(5)
```

```python
states = set(df['STATES'])
for state in states:
    print(state)
    state_wise_data = df[df['STATES']==state]
    print(state_wise_data)
    sns.lineplot(x='YEAR',y='TOTAL CRIMES AGAINST WOMEN',data=state_wise_data,marker='x')
    plt.show()
```

```python
states = set(df['STATES'])
for state in states:
    print(state)
    state_wise_data = df[df['STATES']==state]
    sns.lineplot(x='YEAR',y='RAPE',data=state_wise_data,marker='x')
    plt.title(state)
    plt.show()
```

```python
df.head(0)
```

```python
print("Average number of crimes happening in india per year")
year_wise_average_crimes = df.groupby('YEAR')['TOTAL CRIMES AGAINST WOMEN'].mean()
year_wise_average_crimes.plot(kind='line',ylabel='Average number of crimes',title='Average')
plt.show()
year_wise_average_crimes.plot(kind='bar',ylabel='Average number of crimes',title='Average')
plt.show()
```

```python
df.head(5)
```

```python
for state in states:
    state_wise_dowry_deaths = df[df['STATES']==state]
    print(state)
    sns.lineplot(x='YEAR',y='DOWRY DEATHS',data=state_wise_dowry_deaths)
    plt.show()
```

```python
df.head(5)
```

```python
print("Number of sexual harrasments in each for each year : ")
for state in states:
    print(state)
    state_wise_sexual_harrasments_data = df[df['STATES']==state]
    sns.barplot(x='YEAR',y='SEXUAL HARRASSMENT',data=state_wise_sexual_harrasments_data)
    plt.show()
```

```python
print("Average number of sexual harrassments happening in india every year : ")
averagenumber_of_sexual_harrasments = df.groupby('YEAR')['SEXUAL HARRASSMENT'].mean()
averagenumber_of_sexual_harrasments.plot(kind='bar',ylabel='Average number of sexual harrassments every year')
```
