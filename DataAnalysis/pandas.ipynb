### Pandas
Pandas is a data manipulation library
->It mainly gives us two data structures one is series and other is data frame
series -> one dimensional array object
dataframe -> two dimensional array which is mutable , heterogenious , and can have labelled rows and columns

Series is something which is similar to column in a table or excel sheet

```python
#Creating series using pandas
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series : ")
print(series)
```

```python
#Creating series from a dictionary'
data = {'66' : "Sahith","67":"Manoj","68":"Ranga"}
series = pd.Series(data)
print("Series : ")
print(series)
```

```python
data = ["Sahith","Manoj","Kanna"]
index = [66,67,68]
series = pd.Series(data,index=index)
print(series)
```

```python
#Data frame
data = {
    'Name' : ["Sahith","Kanna","Pandu"],
    'Age'  : [18,19,20],
    'CGPA' : [9.34,9.33,9.32]
}

df = pd.DataFrame(data,index=[1,2,3])
print(df)
```

```python
df['Sales_filled'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
#Renaming the columns in the dataset
df = df.rename(columns={'Date':'Sale date'})
df.head(5)
```

```python
#Changing the data type
df['New_Values'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head(5)
```

```python
import math
df['Squared_values'] = df['Value'].apply(lambda x:x**2)
df.head()
```

```python
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_aggregates = df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_aggregates
```

```python
df1 = pd.DataFrame(
    {
    'Name' : ["Sahith","Pandu","Kanna"],
    'Roll' : [66,67,68]
    })
df2 = pd.DataFrame(
    {'Name' : ["Rahul","Kanna","Krishna"],
    'Roll' : [69,70,71]}
)
```
