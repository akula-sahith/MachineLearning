### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#basic plotting
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Basic")
plt.show()
```

```python
#Customized plot
plt.plot(x,y,color='red',linestyle='--',marker='o')
```

```python
plt.grid(True)
plt.show()
```

```python
#Subplotts

x = list(range(1,6))
y = [i**2 for i in x]
z = [i**3 for i in x]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y,color='red')
plt.title("Plot 1")


plt.subplot(2,2,2)
plt.plot(y,x,color='green')
plt.title("Plot 2")


plt.subplot(2,2,3)
plt.plot(x,y,color='blue')
plt.title("Plot 3")


plt.subplot(2,2,4)
plt.plot(x,y,color='orange')
plt.title("Plot 4")
```

##### BAR GRAPHS

```python
categories = ['A','B','C','D','E']
values = [10 , 5 , 6 , 3 , 4]
plt.bar(categories,values,color='blue')
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Bar graph")
# plt.grid(True)
plt.show()
```

### HISTOGRAMS

```python
numbers = list(range(1,7))
values = [2,4,9,16,25,36]
plt.figure(figsize=(9,11))

plt.subplot(2,2,1)
plt.hist(values,bins=5,color='blue',edgecolor='black')
plt.title("Histogram")

plt.subplot(2,2,2)
plt.plot(numbers,values)
plt.title("Normal Plotting")

plt.subplot(2,2,3)
plt.scatter(numbers,values,marker='x')
plt.title("Scatter plot")

plt.subplot(2,2,4)
explodes = [0,0,0,0,0,0.4]
plt.pie(values,explode=explodes,labels=numbers)
plt.title("Pie Chart")
```
