```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data.csv')
df
```

```python
df.head(10)
```

```python
product_category_revenue = df.groupby('Product Category')['Total Revenue'].sum()
product_category_revenue.plot(kind='bar',color='green')
```

```python
explodes = [0 for i in range(len(product_category_revenue))]
k = 0
for i in product_category_revenue:
    if i == max(product_category_revenue):
        explodes[k] = 0.2
    else:
        explodes[k] = 0
    k += 1
product_category_revenue.plot(kind='pie',autopct='%1.1f%%',shadow=True)
```

```python
payment_type_total_revenue = df.groupby('Payment Method')['Total Revenue'].sum()
payment_type_total_revenue.plot(kind='bar',grid=True)
print(payment_type_total_revenue)
```

```python
product_category_revenue = df.groupby('Product Category')['Total Revenue'].agg(['mean','sum'])
product_category_revenue.plot(kind='bar',xlabel='Product',ylabel=['mean','sum'],color=['blue','orange'])
```

```python
revenue_by_region = df.groupby('Region')['Total Revenue'].sum()
revenue_by_region.plot(kind='bar',color='red')
```
